```python
import morph_kgc
```

```python
var_list = [1, 2, 3, 4, 5, 17]

# Template configuration with placeholder for variable substitution
config_template = """[GTFS_CSV]
mappings: mappings/var{var}_mapping.yml
"""

# List to hold the generated configs
config_list = [config_template.format(var=var) for var in var_list]
```

```python
for idx,config in enumerate(config_list):

    ## Serialize the graph
    g = morph_kgc.materialize(config)
    turtle = g.serialize(format="turtle")

    with open(rf'variable_ttl\output{idx}.ttl', 'w') as file:
        file.write(turtle)
```
