```python
import numpy as np
import pandas as pd
from datetime import datetime
from pathlib import Path   
import glob
import os
```

```python
# 1) Point this at the directory containing your timestamped files
data_dir = "/Users/rastegar-a/Documents/GitHub/i-adopt-llm-based-service/benchmarking_outputs/T=0"

# 2) Grab all files matching the pattern
pattern = os.path.join(data_dir, "iadopt_metrics*.xlsx")
files = glob.glob(pattern)

# 3) Read and concatenate
dfs = []
for fn in files:
    df = pd.read_excel(fn, sheet_name="summary")
    dfs.append(df)
all_data = pd.concat(dfs, ignore_index=True)

# 4) Compute the mean and std of F_exact per Prompt × Model
agg_data = (
    all_data
    .groupby(['Shot', 'Model'], as_index=False)['F_close']
    .agg(mean='mean', std='std')
)

# 5) Sort descending by mean F_exact
agg_data = agg_data.sort_values(by='mean', ascending=False).reset_index(drop=True)

# Display
print("T=0:\n")
display(agg_data)
```
